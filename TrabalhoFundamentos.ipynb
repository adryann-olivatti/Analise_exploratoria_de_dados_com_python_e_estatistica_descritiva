```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from IPython.display import HTML, display
```

```python

#importando a tabela do github da Cibele para fazer o tratamento dos dados
url = "https://raw.githubusercontent.com/cibelerusso/Datasets/main/DadoseDecisoes.csv"
dados = pd.read_csv(url)

# a) Inicialmente, classifique os funcionários em dois grupos (Produtividade bin: ”Alta”ou ”Baixa”):
# • Alta produtividade: funcionários com produtividade acima da mediana;
# • Baixa produtividade: funcionários com produtividade abaixo ou igual á mediana

mediana_produtividade = dados['Produtividade'].median() #calculando a mediana dos dados

dados['Produtividade_bin'] = np.where(
    dados['Produtividade'] > mediana_produtividade, 'Alta', 'Baixa'
    ) #criando a coluna com os dados binarizados em produtividade 'Baixa e Alta'
print(f"A mediana da produtividade é {mediana_produtividade}")
```

1. **Variáveis Qualitativas Nominais**

São categorias que servem apenas para classificar ou identificar, sem nenhuma relação de ordem ou hierarquia.

**ID**

Motivo: É apenas um código numérico para diferenciar um funcionário do outro. Esses números não indicam quantidade e não faz sentido somá-los ou tirar a média deles.

**Gênero**

Motivo: Divide os dados em grupos (ex: Masculino, Feminino). São apenas rótulos textuais e nenhum grupo é "maior" ou "melhor" que o outro.

**Departamento**

Motivo: Classifica o funcionário pelo seu setor de trabalho (ex: TI, RH, Vendas). São categorias puramente qualitativas e sem hierarquia.

**Home_Office**

Motivo: É uma variável categórica binária (geralmente respondida com "Sim" ou "Não"). Apenas indica a modalidade de trabalho, sem valor numérico real.

2. **Variáveis Qualitativas Ordinais**

São categorias (rótulos ou conceitos), mas que possuem uma ordem ou escala natural de intensidade.

**Satisfação**

Motivo: Mede um sentimento ou percepção em graus (ex: Insatisfeito, Neutro, Satisfeito). Embora às vezes receba notas de 1 a 5, os números servem apenas para representar uma ordem clara de preferência ou contentamento.

3. **Variáveis Quantitativas Discretas**

São valores numéricos que representam uma contagem, ou seja, assumem apenas valores inteiros e isolados (não existem valores "quebrados").

**Idade**

Motivo: Embora o tempo corra continuamente, a idade em bancos de dados de empresas é registrada como uma contagem de anos completos vividos (25 anos, 36 anos, etc.).

**Cursos_Realizados**

Motivo: Resulta de uma contagem direta de eventos. O funcionário faz 1, 2 ou 3 cursos; é impossível ele ter feito "2,4 cursos".

4. **Variáveis Quantitativas Contínuas**

São valores numéricos que resultam de uma medição e podem assumir qualquer valor decimal (números "quebrados") dentro de um intervalo.

**Salário**

Motivo: É um valor monetário medido. O dinheiro pode ser infinitamente fracionado em centavos (ex: R$ 4.500,50), o que o torna uma variável contínua.

**Horas_Trabalhadas**

Motivo: É uma grandeza de tempo. O tempo é contínuo por natureza, pois um funcionário pode trabalhar frações de horas (ex: 40 horas e 30 minutos equivalem a 40,5 horas).

**Tempo_Empresa**

Motivo: Assim como as horas trabalhadas, mede a duração da jornada do funcionário na empresa (pode ser medido em meses ou anos parciais, como 2,5 anos de casa).

**Produtividade**

Motivo: Geralmente é calculada através de uma taxa, índice ou porcentagem (ex: 87,5% das metas batidas). Como o resultado dessa divisão gera números decimais contínuos, ela se enquadra aqui.



```python
#(c) Para variáveis qualitativas, utilize tabelas de frequência, gráficos de barras ou gráficos de setores;
# --- PREPARAÇÃO EXTRA DOS SEUS DADOS ---
# 1. Ajustando a Satisfação de números para categorias (Baixa, Média, Alta)
dados["Satisfação_Cat"] = pd.cut(
    dados["Satisfação"],
    bins=[0, 75, 90, 101],
    labels=["Baixa", "Média", "Alta"],
)

# Lista com as variáveis que você quer analisar
variaveis = ["ID", "Gênero", "Departamento", "Home_Office", "Satisfação_Cat"]

# --- LAÇO PARA GERAR TABELA E GRÁFICO LADO A LADO ---
for var in variaveis:
    # Ajuste para o ID não quebrar o visual (mostra apenas os 10 primeiros)
    if var == "ID":
        dados_var = dados["ID"].head(10)
        titulo_sufixo = " (Primeiros 10)"
    else:
        dados_var = dados[var]
        titulo_sufixo = ""

    # 1. Calcular Frequência Absoluta e Relativa
    freq_abs = dados_var.value_counts()
    freq_rel = (dados_var.value_counts(normalize=True) * 100).round(1)

    tabela_frequencia = pd.DataFrame(
        {"Frequência Absoluta": freq_abs, "Frequência Relativa (%)": freq_rel}
    )

    # 2. Criar a estrutura com 2 colunas lado a lado
    fig, axes = plt.subplots(1, 2, figsize=(15, 4))

    # --- LADO ESQUERDO: Desenha a Tabela ---
    axes[0].axis("off")  # Limpa o fundo de gráfico
    tabela_visual = axes[0].table(
        cellText=tabela_frequencia.values,
        rowLabels=tabela_frequencia.index,
        colLabels=tabela_frequencia.columns,
        loc="center",
        cellLoc="center",
    )
    tabela_visual.scale(1, 1.8)  # Ajusta o tamanho das linhas
    axes[0].set_title(
        f"Tabela de Frequência: {var.replace('_Cat','')}{titulo_sufixo}",
        fontsize=13,
        fontweight="bold",
        pad=10,
    )

    # --- LADO DIREITO: Desenha o Gráfico de Barras ---
    dados_var.value_counts().plot(kind="bar", ax=axes[1], color="#5da4e2")
    axes[1].set_title(
        f"Gráfico de Barras: {var.replace('_Cat','')}{titulo_sufixo}",
        fontsize=13,
        fontweight="bold",
    )
    axes[1].set_xlabel("Categorias")
    axes[1].set_ylabel("Quantidade")
    axes[1].tick_params(
        axis="x", rotation=0 if var != "Departamento" else 30
    )  # Inclina se for o setor

    # Ajusta os espaços e plota o par (Tabela + Gráfico)
    plt.tight_layout()
    plt.show()
    print("\n" + "-" * 100 + "\n")
```

```python
# (d) Para variáveis quantitativas, utilize medidas descritivas e visualizaçõess adequadas (como his-
# togramas, boxplots, gráficos de dispersão ou gráficos de violino);
# Configuração visual básica
sns.set_theme(style="whitegrid")

# Lista com as suas variáveis quantitativas
variaveis_quant = [
    "Idade",
    "Cursos_Realizados",
    "Salário",
    "Horas_Trabalhadas",
    "Tempo_Empresa",
    "Produtividade",
]

# --- LAÇO PARA GERAR MEDIDAS E OS 3 GRÁFICOS LADO A LADO ---
for var in variaveis_quant:
    print(f"======= ANÁLISE DA VARIÁVEL: {var} =======")

    # 1. Calcular as Medidas Descritivas
    medidas = dados[var].describe()
    mediana_valor = dados[var].median()

    # Criando a tabela limpa
    tabela_medidas = pd.DataFrame(
        {
            "Média": [round(medidas["mean"], 2)],
            "Mediana": [round(mediana_valor, 2)],
            "Desvio Padrão": [round(medidas["std"], 2)],
            "Mínimo": [round(medidas["min"], 2)],
            "Máximo": [round(medidas["max"], 2)],
        },
        index=["Valores"],
    )

    # Exibe a tabela de texto no console
    display(tabela_medidas)

    # 2. Configurar a tela para colocar TRÊS gráficos lado a lado
    fig, axes = plt.subplots(1, 3, figsize=(22, 5))

    # --- GRAFICO 1 (Esquerda): Histograma ---
    sns.histplot(data=dados, x=var, kde=True, ax=axes[0], color="#5da4e2")
    axes[0].set_title(f"Histograma de {var}", fontsize=12, fontweight="bold")
    axes[0].set_xlabel(var)
    axes[0].set_ylabel("Frequência")

    # --- GRAFICO 2 (Meio): Boxplot ---
    sns.boxplot(data=dados, x=var, ax=axes[1], color="#99ff99")
    axes[1].set_title(f"Boxplot de {var}", fontsize=12, fontweight="bold")
    axes[1].set_xlabel(var)

    # --- GRAFICO 3 (Direita): Gráfico de Dispersão ---
    # Se a variável for a própria Produtividade, cruzamos com o Salário para não ficar uma linha reta
    var_concorrente = "Salário" if var == "Produtividade" else "Produtividade"

    sns.scatterplot(
        data=dados, x=var, y=var_concorrente, ax=axes[2], color="#ec7063"
    )
    axes[2].set_title(
        f"Dispersão: {var} vs {var_concorrente}",
        fontsize=12,
        fontweight="bold",
    )
    axes[2].set_xlabel(var)
    axes[2].set_ylabel(var_concorrente)

    # Mostra o conjunto de gráficos na tela
    plt.tight_layout()
    plt.show()
    print("\n" + "-" * 110 + "\n")
```

```python

# 1. Configuração visual do Seaborn
sns.set_theme(style="whitegrid")

# Separando as variáveis por tipo para fazer as análises corretas
lista_qualitativas = ["Gênero", "Departamento", "Home_Office"]
lista_quantitativas = [
    "Idade",
    "Salário",
    "Horas_Trabalhadas",
    "Tempo_Empresa",
    "Cursos_Realizados",
]

# =====================================================================
# PARTE 1: ASSOCIAÇÃO COM VARIÁVEIS QUALITATIVAS (Cruzamento de categorias)
# =====================================================================
print(
    "=== 1. ASSOCIAÇÕES ENTRE PRODUTIVIDADE BIN E VARIÁVEIS QUALITATIVAS ===\n"
)

for var in lista_qualitativas:
    print(f"--> Cruzamento: Produtividade_bin vs {var}")

    # Cria a tabela de contingência (cruzamento) em percentual para comparar melhor
    tabela_cruzada = pd.crosstab(
        dados[var], dados["Produtividade_bin"], normalize="index"
    )
    tabela_cruzada = (tabela_cruzada * 100).round(1)

    # Configura a exibição lado a lado (Tabela e Gráfico)
    fig, axes = plt.subplots(1, 2, figsize=(14, 4))

    # Desenha a tabela no lado esquerdo
    axes[0].axis("off")
    tabela_visual = axes[0].table(
        cellText=tabela_cruzada.values,
        rowLabels=tabela_cruzada.index,
        colLabels=tabela_cruzada.columns,
        loc="center",
        cellLoc="center",
    )
    tabela_visual.scale(1, 1.8)
    axes[0].set_title(
        f"Distribuição % de {var}", fontsize=12, fontweight="bold"
        )

    # Desenha o gráfico de barras agrupadas no lado direito
    sns.countplot(
        data=dados,
        x=var,
        hue="Produtividade_bin",
        ax=axes[1],
        palette="coolwarm",
    )
    axes[1].set_title(f"Gráfico: {var} por Produtividade")
    axes[1].set_ylabel("Quantidade de Funcionários")
    axes[1].set_xlabel(var)

    plt.tight_layout()
    plt.show()
    print("\n" + "-" * 100 + "\n")


# =====================================================================
# PARTE 2: ASSOCIAÇÃO COM VARIÁVEIS QUANTITATIVAS (Comparação de médias/grupos)
# =====================================================================
print(
    "=== 2. ASSOCIAÇÕES ENTRE PRODUTIVIDADE BIN E VARIÁVEIS QUANTITATIVAS ===\n"
)

for var in lista_quantitativas:
    print(
        f"--> Análise: {var} por Grupo de Produtividade"
        )

    # Cria a tabela resumo comparando a Média e Mediana da variável para quem é Alta ou Baixa produtividade
    tabela_resumo = dados.groupby(
        "Produtividade_bin"
        )[var].agg(["mean", "median"]).round(2)
    tabela_resumo.columns = ["Média", "Mediana"]

    # Configura a exibição lado a lado (Tabela e Boxplot)
    fig, axes = plt.subplots(1, 2, figsize=(14, 4))

    # Desenha a tabela resumo no lado esquerdo
    axes[0].axis("off")
    tabela_visual = axes[0].table(
        cellText=tabela_resumo.values,
        rowLabels=tabela_resumo.index,
        colLabels=tabela_resumo.columns,
        loc="center",
        cellLoc="center",
    )
    tabela_visual.scale(1, 1.8)
    axes[0].set_title(f"Resumo de {var}", fontsize=12, fontweight="bold")

    # Desenha o Boxplot comparativo no lado direito (perfeito para ver diferença de comportamento)
    sns.boxplot(
        data=dados,
        x="Produtividade_bin",
        y=var,
        ax=axes[1],
        palette="coolwarm",
        hue="Produtividade_bin",
        legend=False,
    )
    axes[1].set_title(f"Boxplot de {var} por Produtividade")
    axes[1].set_xlabel("Grupo de Produtividade")
    axes[1].set_ylabel(var)

    plt.tight_layout()
    plt.show()
    print("\n" + "-" * 100 + "\n")
```

**h) Discuta possíveis limitações dessa análise exploratória, destacando por que associações observadas não necessariamente possuem significância estatística. Indique quais seriam as análises
complementares para a análise confirmatória e quais seriam as suas implicações**

A análise exploratória de dados (AED) é uma etapa fundamental, mas possui limitações cruciais que impedem a tomada de decisões definitivas:

Efeito do Acaso (Flutuação Amostral): Os padrões e diferenças visuais que observamos nos gráficos deste conjunto de  funcionários podem ser apenas uma coincidência  da amostra coletada. Sem testes matemáticos, não podemos garantir que esses mesmos padrões se repetiriam se analisássemos toda a empresa ou novos contratados.

Correlação não implica Causalidade: O fato de haver uma forte associação entre duas variáveis  não prova que uma é a causa direta da outra. É possível, por exemplo, que funcionários que já são naturalmente mais produtivos acabem assumindo mais demandas e fiquem mais tempo na empresa, invertendo a lógica da causa.

Fatores de Confusão : Uma associação que parece forte visualmente pode ser explicada por uma terceira variável que não foi isolada. Por exemplo, o departamento de TI pode parecer mais produtivo apenas porque os funcionários de lá realizam mais cursos, e não pelo departamento em si.

**Para migrar de uma análise descritiva ("o que os dados mostram") para uma análise confirmatória ("o que a matemática prova"), é necessário aplicar a Estatística Inferencial (fazer uma análise confirmatória).**
